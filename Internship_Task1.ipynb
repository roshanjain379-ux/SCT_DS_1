```python
RR
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv("/content/API_SP.POP.TOTL_DS2_en_csv_v2_744461.csv", skiprows=4)


india_data = data[data["Country Name"] == "India"]


years = india_data.columns[4:-1]r
population = india_data.iloc[0, 4:-1].astype(float)

plt.figure(figsize=(12,6))
plt.bar(years, population/1e9, color="royalblue")
plt.title("India's Population Growth (19602022)")
plt.xlabel("Year")
plt.ylabel("Population (in Billions)")
plt.show()


age_groups = ["0-14 years", "15-64 years", "65+ years"]
percentages = [25.7, 67.3, 7.0]

plt.figure(figsize=(7,5))
plt.bar(age_groups, percentages, color=["gold", "skyblue", "lightcoral"])
plt.title("India's Population Distribution by Age (2022)")
plt.ylabel("Percentage of Total Population")
plt.show()
```
